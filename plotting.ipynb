```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
colors = ['royalblue', 'violet', 'crimson', 'limegreen', 'sienna', 'orange', 'darkviolet', 'black']
```

```python
ts_data = np.load("/home/ubuntu/nums/outputs/cholesky-ts-64.npy")
#rows,cols,b,mult_time,inv_time,total_time
ts_data.shape
```

```python
plt.figure(figsize=(10,5))
for blk in range(7,11):
    b = 2 ** blk
    blk_cur = ts_data[ts_data[:,2]==b]
    plt.semilogy(blk_cur[:,1],blk_cur[:,5],label="Block Size:{}".format(b))
plt.title("Time vs Number of Features",fontsize=20)
plt.ylabel("Time(s)",fontsize=20)
plt.xlabel("Number of Features",fontsize=20)
plt.legend()
```

```python
pd_data = np.load("/home/ubuntu/nums/outputs/cholesky-pd-big-64.npy")
#rows,b,inv_time
pd_data.shape
```

```python
# ts_data[:,2]
plt.figure(figsize=(10,5))
for blk in range(11,20):
    b = 2 ** blk
    blk_cur = pd_data[pd_data[:,1]==b]
    plt.loglog(blk_cur[:,0],blk_cur[:,2],label="Block Size:2^{}".format(blk))
plt.title("Time vs Size of X",fontsize=20)
plt.ylabel("Time(s)",fontsize=20)
plt.xlabel("Size of X",fontsize=20)
plt.legend()
```
